```
#####################################################################
#-------------frequent terms of top10 icd9 category-----------------#
#####################################################################

frequent_category=['401','427','276','414','272','250','428','518','285','584']
category_med_terms={}

#common terms that are common to the ICD categories
common_diagnosis_terms=["monday","phone:<telephone"]

while True:
  
  for i in frequent_category:
      diagnosis_x = diagnosis_icd9category.loc[diagnosis_icd9category.icd9_category.apply(lambda x: i in x)].copy()
      diagnosis_x['medical_terms_list'] = diagnosis_x['medical_terms_list'].apply(literal_eval)
      medterms_list = diagnosis_x.medical_terms_list.to_list()    
      allmedterms = [ term for item in medterms_list for term in list(set(item)) ]
      
      dict_med = dict(sorted(dict(Counter(allmedterms)).items(), key = lambda x: x[1], reverse = True))
      
      dict_med_list = list(dict_med)
      dict_med_list_notcommon=[a for a in dict_med_list if a not in common_diagnosis_terms]
      
      category_med_terms[i]=dict_med_list_notcommon[:10]


  diagnosis_icdcategory_medterms={'icd9category':[i for i in category_med_terms],
                                  'frequent_medterms':[category_med_terms[i] for i in category_med_terms]}
  diagnosis_icdcategory_medterms_df=pd.DataFrame.from_dict(diagnosis_icdcategory_medterms)
  display(diagnosis_icdcategory_medterms_df)
  
  # Adding terms that are common in 6 of the terms to common_diagnosis_terms
  y=diagnosis_icdcategory_medterms_df.frequent_medterms.to_list()
  y1=[a for b in y for a in b]
  medical=dict(Counter(y1))
  if [k for (k,v) in medical.items() if v>=6]:
    common_diagnosis_terms.extend( [k for (k,v) in medical.items() if v>=6] )
  else:
    break
```

```
#################################################
#-----------weights of each category------------#
#################################################
def predict_category(x):
  weight=[1.0,0.9,0.8,0.7,0.6,0.5,0.4,0.3,0.2,0.1] # Weights of terms 
  predict_category_wt={}
  for i in category_med_terms:
    index_binary=[1 if m in x else 0 for m in category_med_terms[i]]
    weighted_predict=sum([m*n for m,n in zip(weight,index_binary)])
    predict_category_wt[i]=weighted_predict
  return predict_category_wt

```

```
#################################################################
#--------------------predicted categories-----------------------#
#################################################################
def predictbasedonwts(x):
  x1=dict(sorted(x.items(),key= lambda y: y[1],reverse=True))
  x2={k:v for (k,v) in x1.items() if v>0}
  repeatedval=dict(Counter(list(x2.values())))
  repeatedvalgreaterthan1=[k for (k,v) in repeatedval.items() if v>1]
  x3=dict(list(x2.items())[:10])
  x4=[v for (k,v) in x3.items() if v in repeatedvalgreaterthan1]
  x5={k:v for (k,v) in x2.items() if v in x4}
  x3.update(x5)
  return list(x3.keys())


  

```

```
def predict_code(x):
  weight=[1.0,0.9,0.8,0.7,0.6,0.5,0.4,0.3,0.2,0.1]
  predict_code_wt={}
  for i in code_med_terms:
    index_binary=[1 if m in x else 0 for m in code_med_terms[i]]
    weighted_predict=sum([m*n for m,n in zip(weight,index_binary)])
    predict_code_wt[i]=weighted_predict
  return predict_code_wt
```

```
predict_diagnosis_icd9code['predict_icd_weights']=predict_diagnosis_icd9code.medical_terms_list.progress_apply(lambda x:predict_code(x))
```

```
#####################################################################
#-------------frequent terms of top10 icd9 codes-----------------#
#####################################################################


code_med_terms={}

#common terms that are common to the ICD categories
common_diagnosis_terms=["monday","phone:<telephone"]

while True:
  
  for i in frequent_icd10codes:
      diagnosis_x = diagnosis_terms_codes.loc[diagnosis_terms_codes.ICD10.apply(lambda x: i in x)].copy()
      diagnosis_x['medical_terms_list'] = diagnosis_x['medical_terms_list'].apply(literal_eval)
      medterms_list = diagnosis_x.medical_terms_list.to_list()    
      allmedterms = [ term for item in medterms_list for term in list(set(item)) ]
      
      dict_med = dict(sorted(dict(Counter(allmedterms)).items(), key = lambda x: x[1], reverse = True))
      
      dict_med_list = list(dict_med)
      dict_med_list_notcommon=[a for a in dict_med_list if a not in common_diagnosis_terms]
      
      code_med_terms[i]=dict_med_list_notcommon[:10]


  diagnosis_icd10code_medterms={'icd10code':[i for i in code_med_terms],
                                  'frequent_medterms':[code_med_terms[i] for i in code_med_terms]}
  diagnosis_icd10code_medterms_df=pd.DataFrame.from_dict(diagnosis_icd10code_medterms)
  display(diagnosis_icd10code_medterms_df)
  
  # Adding terms that are common in 6 of the terms to common_diagnosis_terms
  y=diagnosis_icd10code_medterms_df.frequent_medterms.to_list()
  y1=[a for b in y for a in b]
  medical=dict(Counter(y1))
  if [k for (k,v) in medical.items() if v>=6]:
    common_diagnosis_terms.extend( [k for (k,v) in medical.items() if v>=6] )
  else:
    break
```
